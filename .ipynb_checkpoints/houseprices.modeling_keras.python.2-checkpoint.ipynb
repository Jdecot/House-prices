```python
import keras
from keras.losses import mean_squared_error
from keras.optimizers import SGD
from keras.layers import Dense
from keras.layers import Dropout
from keras.models import Sequential

from keras import metrics

sgd = SGD(lr=0.01, 
         decay=1e-6, 
         momentum=0.9)

x_train = train_rwrk[modelisation_feature].to_numpy()
y_train = train_rwrk[y].to_numpy()
x_test = test_rwrk[modelisation_feature].to_numpy()
y_test = test_rwrk[y].to_numpy()

x_train = x_train.astype('float32')
y_train = y_train.astype('float32')
x_test = x_test.astype('float32')
y_test = y_test.astype('float32')
```

```python
x_train.shape[1]
```

```python
# https://www.kaggle.com/hugosjoberg/house-prices-prediction-using-keras
import tensorflow as tf
from tensorflow.keras import layers
```

```python
model = Sequential()
model.add(Dense(81, input_dim=81, kernel_initializer='normal', activation='relu'))
model.add(Dense(200, kernel_initializer='normal', activation='relu'))
model.add(Dense(100, kernel_initializer='normal', activation='relu'))
model.add(Dense(50, kernel_initializer='normal', activation='relu'))
model.add(Dense(25, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))

model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adadelta())


model.fit(x_train, y_train, epochs=100, batch_size=10)    
```
